## 3. Adding Legends (Multiple Lines on One Plot)

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1 , 2 , 3 , 4 ,5 ]
y = [10 ,20,25,30,50]
z = [5,15,20,25,40]
```

```python
plt.plot(x , y, label= "Line 1" , color= "blue",linestyle = "--")
plt.plot(x , z , label = "Line 2", color = "black" , linestyle = ":")

plt.title("Multiple Lines on one plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() # Displays a legend to differentiate multiple lines.
plt.show()
```

# 3️⃣ Bar Charts & Scatter Plots in Matplotlib

## 1️⃣ Bar Charts – Visualizing Categorical Data

```python
import matplotlib.pyplot as plt
```

```python
x = [ "A" , "B", "C" , "D", "F" ]
y = [10 ,20,30,40,50]
```

```python
plt.bar(x , y , color = "blue") # Creates a vertical bar chart.

plt.title("Basic Bar chart")
plt.xlabel("---------> x <---------")
plt.ylabel("--------- > y < ---------")
plt.show()
```

## 📌 Horizontal Bar Chart

```python
plt.barh(x, y, color="green")  # 'barh' creates horizontal bars

plt.title("Horizontal Bar Chart")
plt.xlabel("---------> y <---------")
plt.ylabel("---------> x <---------")

plt.show()
```

## 📌 Grouped Bar Chart (Comparing Two Sets of Data)

```python
import numpy as np
```

```python
x = ["A", "B", "C", "D", "E"]
y =  [10, 20, 15, 25, 30]
z = [12, 18, 22, 20, 27]
```

```python
x1 = np.arange(len(x)) # Get numerical position ## Uses np.arange() for positioning bars side by side
width = 0.3 #Bar width

# plotting bars
plt.bar(x1 - width/2, y , width , label = "Set 1", color = "green")
plt.bar(x1 + width / 2 , z , width , label="Set 2" , color = "orange")

# Customizing labels
plt.xticks(x1, x)  # Set x-axis labels ##plt.xticks(x, categories) keeps proper labels.
plt.title("Grouped Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()

plt.show()
```

## 2️⃣ Scatter Plots – Visualizing Relationships Between Variables

```python
import matplotlib.pyplot as plt

# Data
x = [10, 20, 30, 40, 50, 60, 70]
y = [5, 15, 25, 20, 35, 45, 55]

# Creating a scatter plot
plt.scatter(x, y, color="red", marker="^", s=100)  # 's' controls marker size 
#plt.scatter(x, y) → Creates scatter plot.

# Adding title and labels
plt.title("Basic Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

## Scatter Plot with Categories (Color Coding)

```python
import numpy as np

# Data
x = np.random.rand(50) * 100  # 50 random points
y = np.random.rand(50) * 100
colors = np.random.rand(50)  # Assign random colors
sizes = np.random.rand(50) * 300  # Assign random sizes

# Scatter plot with colors and sizes
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap="viridis") #alpha=0.6 → Adds transparency.

# Adding title and labels
plt.title("Scatter Plot with Color Mapping")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()  # Adds a color scale

plt.show()
```

# 4️⃣Subplots Multiple Plots in One Figure

## 1️⃣ Creating Basic Subplots (Using plt.subplot())

```python
import matplotlib.pyplot as plt

# First subplot
plt.subplot(2, 2, 1)  # (rows=2, cols=2, position=1)
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Line Plot")

# Second subplot # plt.subplot() adds a new plot.
plt.subplot(2, 2, 2)  # Position 2
plt.bar(["A", "B", "C"], [5, 7, 3])
plt.title("Bar Chart")

# Third subplot
plt.subplot(2, 2, 3)  # Position 3
plt.scatter([1, 2, 3], [30, 10, 20])
plt.title("Scatter Plot")

# Fourth subplot
plt.subplot(2, 2, 4)  # Position 4
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=4)
plt.title("Histogram")

plt.tight_layout()  # Adjust spacing
plt.show()
```

## 2️⃣ Creating Flexible Subplots (Using plt.subplots())

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Line plot
axes[0, 0].plot(np.random.rand(10))
axes[0, 0].set_title("Line Plot")

# Bar chart
axes[0, 1].bar(["A", "B", "C"], [5, 7, 3], color="green")
axes[0, 1].set_title("Bar Chart")

# Scatter plot
axes[1, 0].scatter(np.random.rand(10), np.random.rand(10), color="red")
axes[1, 0].set_title("Scatter Plot")

# Histogram
axes[1, 1].hist(np.random.randn(100), bins=10, color="purple")
axes[1, 1].set_title("Histogram")

plt.tight_layout()  # Adjust spacing
plt.show()
```

# 5️⃣ Working with Images & Heatmaps in Matplotlib

## 1️⃣ Displaying Images with plt.imshow()

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a random 10x10 image (grayscale)
image = np.random.rand(10, 10)

plt.imshow(image, cmap="gray")  # 'gray' colormap for grayscale images
plt.title("Random Grayscale Image")
plt.colorbar()  # Shows value scale
plt.show()
# plt.imshow(image) → Displays a 2D array as an image.
#cmap="gray" → Colormap for grayscale images. You can also try cmap="hot", cmap="viridis", cmap="jet".
#plt.colorbar() → Adds a color scale beside the image.
```

## 📌 Displaying Color Images (RGB)

```python
import matplotlib.pyplot as plt
import numpy as np

# Random RGB Image (10x10 pixels)
image = np.random.rand(10, 10, 3)

plt.imshow(image)
plt.title("Random RGB Image")
plt.show()
#(height, width, 3) → RGB image.
```

## 2️⃣ Visualizing MNIST Dataset Example (28x28 Digits)

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load handwritten digits dataset (similar to MNIST)
digits = load_digits()

# Display the first digit
plt.imshow(digits.images[0], cmap="gray")
plt.title(f"Digit Label: {digits.target[0]}")
plt.colorbar()
plt.show()
#Each digit is 8x8 pixels in this dataset. MNIST is 28x28 pixels.
#digits.images[0] → Returns the pixel values of the first image.
#digits.target[0] → Returns the label (actual number).
```

##  3️⃣ Heatmaps (Useful for Confusion Matrices & Correlations)

```python
# 📌 Creating a Heatmap (Manually)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Random 5x5 data (like a confusion matrix)
data = np.random.randint(1, 100, (5, 5))

plt.imshow(data, cmap="coolwarm", interpolation="nearest")
plt.title("Heatmap Example")
plt.colorbar()
plt.show()
# interpolation="nearest" → Keeps the blocks sharp.
#cmap="coolwarm" → Try other colormaps like plasma, hot, viridis
```

## 📌 Plotting Confusion Matrix (Example Visualization in ML)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Example prediction results
y_true = [1, 0, 1, 1, 0, 1, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1, 1, 1]

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Visualize it
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
#✅ Why Heatmaps Matter in ML:

#Confusion Matrices → Understand how well your classifier is performing.
#Correlation Matrices → Analyze how features relate to each other.
```
